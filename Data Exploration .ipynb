```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = np.load('mnist.npz') 
X_train, y_train = data['x_train'], data['y_train']
X_test, y_test = data['x_test'], data['y_test']

# Check the shapes
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
```

```python
# plot 4 images as gray scale
plt.subplot(221)
plt.imshow(X_train[0], cmap=plt.get_cmap('gray'))
plt.subplot(222)
plt.imshow(X_train[1], cmap=plt.get_cmap('gray'))
plt.subplot(223)
plt.imshow(X_train[2], cmap=plt.get_cmap('gray'))
plt.subplot(224)
plt.imshow(X_train[3], cmap=plt.get_cmap('gray'))
# show the plot
plt.show()
```

```python
##Check Data Distribution: Analyze the distribution of labels to ensure the dataset is balanced.


# Count the number of occurrences of each label
unique, counts = np.unique(y_train, return_counts=True)
plt.bar(unique, counts)
plt.title("Distribution of Labels in Training Set")
plt.xlabel("Digit")
plt.ylabel("Frequency")
plt.show()
```

```python
## Inspect Data Statistics: Look at some key statistics to understand the dataset values.

print("Min pixel value:", X_train.min())
print("Max pixel value:", X_train.max())
print("Shape of each image:", X_train[0].shape)
```
